```python
import cfpb_analysis as cfpb
from ipywidgets import fixed, interact
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = cfpb.top_100_companies('http://files.consumerfinance.gov/ccdb/complaints.csv.zip',
                'complaints.csv', -1)
```

```python
@interact
def line_plot(column = fixed('company'), company = sorted(df['company'].unique())):
    
    wordcloud = WordCloud(width=800, height=400).generate(" ".join(df[df['company'] == company]['complaints_description']))
    
    plt.figure(figsize=(12,6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    df[df['company'] == company].set_index('month').plot(title = "COMPLAINTS FOR {} OVER TIME".format(company),
                                                                figsize = (12,6))
```
